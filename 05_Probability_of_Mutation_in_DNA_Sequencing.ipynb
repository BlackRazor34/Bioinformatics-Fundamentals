### Vaka Çalışması: DNA Diziliminde Mutasyon Olasılığı
#### Bir genetik dizileme projesinde, belirli bir gen bölgesindeki 10.000 baz çiftini inceliyoruz (n=10000). Bu bölgedeki doğal arka plan mutasyon oranının her bir baz için binde bir olduğunu biliyoruz (p=0.001). Binom dağılımını kullanarak, bu genetik dizide tam olarak 15 mutasyon görme olasılığımızı veya 10'dan fazla mutasyon (anomali veya hastalık riski) görme olasılığımızı kümülatif olarak hesaplayabiliriz. Bu, makine öğrenmesi modellerinde "anomali tespiti" (anomaly detection) yaparken kullandığımız temel mantıktır.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Biyoinformatik Senaryosu: 10.000 baz çiftlik bir DNA sekansı
n = 10000  # Toplam deneme sayısı (İncelenen baz çifti sayısı)
p = 0.001  # Her bir bazda mutasyon gerçekleşme olasılığı

# İlgilendiğimiz mutasyon sayıları aralığı (0 ile 30 mutasyon arası)
x = np.arange(0, 31)

# Binom Olasılık Kütle Fonksiyonu (PMF) hesaplaması
pmf_mutasyon = binom.pmf(x, n, p)



# Grafiği çizdirme
plt.figure(figsize=(10, 6))
plt.bar(x, pmf_mutasyon, color='#2ca02c', alpha=0.7, edgecolor='black')
plt.title('10.000 Baz Çiftlik DNA Sekansında Beklenen Mutasyon Dağılımı (Binom)')
plt.xlabel('Mutasyon Sayısı (k)')
plt.ylabel('Olasılık ( P(X=k) )')
plt.axvline(x=n*p, color='red', linestyle='--', label=f'Beklenen Değer: {int(n*p)}')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()

# Örnek Bir Çıktı Hesaplama
k = 15
olasilik = binom.pmf(k, n, p)
print(f"Tam olarak {k} mutasyon görülme olasılığı: {olasilik:.4f}")
```
